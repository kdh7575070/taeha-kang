```python
#8.14 누락값처리, 범주형누락값도 처리 *원본 저장이 안된다* 
# (1) dropna는 그냥 다 삭제 (2) fillna는 원하는 값으로 다 채우기 / (3) interpolate는 위아래의 평균값으로 보강

#누락값 처리 (2번)

import numpy as np
import pandas as pd

s = pd.Series([0,1,'0',3])
s #이렇게되면 0,1,3도 object 타입으로 바뀐다

s = pd.Series([0,1,np.nan,3])
s #넘파이의 기본타입이 float. 즉 이렇게되면 0,1,3도 float 타입으로 바뀐다

s.fillna(method='ffill') #값을 넣으면 그 값으로 채워진다. 값대신 특정 속성을 줘볼까 - ffill는 누락값 바로 앞의 값을 누락값으로 치환
s.fillna(method='pad') #마찬가지로 누락값 바로 앞의 값을 누락값으로 치환
s.fillna(method='bfill') #누락값 뒤의 값을 누락값으로 치환

f = pd.read_csv('D:/pandas/실습자료/실습자료/data/flight_route_info.csv', encoding='cp949')

#누락값 보는 방법
f.isnull().sum() #비고라는 컬럼에 누락값이 있다.
f.isnull().sum().sum() #이렇게 하면 행들의 총합 즉 전체의 누락값을 보여준다
f['비고'].value_counts(dropna=False) #비고의 값들의 빈도를 보면 누락값의 빈도가 같이 나온다

#누락값 처리
f['비고'].dtype #열에대한 자료형은 object임
f.fillna("", axis=1, inplace=True) #대체하고자하는 대상(object타입), 적용 하고자 하는 방향, 원본에 작용할지

f['비고'].value_counts(dropna=False) #다시한번 비고의 값들의 빈도를 살펴보면 Nan 대신 ""이 싹다 들어가있다
```

```python
# 범주형 데이터

import pandas as pd
import numpy as np

s = pd.Series(['a','b','c','a'], dtype = "category") #category는 integer, object 이런 타입들의 상위 클래스 - 관리 클래스라고 보면됨
s.shape

s.values.categories #클래스 안에 들어가봤더니 값 자체들은 object다 전체를 관리하는 녀석만 category라고 보면 된다

s.index
s.dtype

try:
#    s[2] = 'd' #d는 카테고리 안에 등록이 안된 값이라서 받아줄 수가 없다
#    s[4] = 'a' #index 를 늘릴 수도 없다
    s[3] = 'c' #index 내에서 요소가 기존에 있을 때만 범주안을 변경할 수 있다
except Exception as e:
    print(e)

cat_1 = pd.Categorical(['a', 'c', 'c', np.nan], categories=['b', 'a', 'c']) #카테고리 만드는 생성자, a,b,c는 범주 안에 넣어준다 
cat_1

df = pd.DataFrame({'cat':cat_1, 's':['a','c','c', np.nan]})
df #두 열은 타입이 다르다!!!! 같이 어떻게 연산 할 수가 없다
df.select_dtypes(exclude=['object']) #오브젝트 타입 열을 제외해라 라고 해보면 카테고리타입만 살아남아 - 원본영향X

df['cat'][3] = 'b' #시리즈 객체의 인덱싱
df #b는 원래 카테고리 안에 있던 값이므로 가능

df.loc[4, :] = ['a', 'a'] #시리즈 객체의 슬라이싱
df #가능
```

```python
#범주형 데이터 누락값 처리

import pandas as pd
import numpy as np

a = pd.read_csv('D:/pandas/실습자료/실습자료/data/airplane_info.csv', encoding='cp949')
a['항공사'].dtype #오브젝트로 들어가있지만 카테고리로 만들면 어떨까? 기존에 없던 항공사가 들어오는 것을 방지 할 수 있어 효율적

air_cat = pd.Categorical(a['항공사']) #Series 객체를 넣으면 자동으로 쉽게 만들어진다

a['항공사'].dtype #원본은 변화없다
air_cat.dtype #카테고리
air_cat.categories #항공사 이름들이 카테고리로 들어갔네
air_cat.shape

a['항공사'] = a['항공사'].astype(air_cat.dtype) #원본 변화시켜주기

a['항공사'].dtype #변경되었다.

try:
    a['항공사'][0]='Air Asia'
except Exception as e:
    print(e)
    
a.isnull().sum()
a['총정원'].isnull().sum()

a['총정원'].dtype #nan 때문에 float 64로 가져온다

a.loc[a['총정원'].isnull(), '총정원'] = a.loc[a['총정원'].isnull(), '정원_Y'] # 총정원열에서 Nan 행을 찾아 정원_Y열의 해당 행값으로 대체 
a.isnull().sum() #DONE
a #정원Y는 int형이지만 float열로 가져오면서 float가 된다
```
